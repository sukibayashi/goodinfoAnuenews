# 分離鉅亨網

```python
import os
import sqlite3
import pandas as pd


# 讀取【鉅亨網】資料庫
def 連接到鉅亨網資料庫():
    conn = sqlite3.connect(os.path.join('file', 'anuenews.db'))
    sql = '''select * from news t'''
    
    df_anuenews = pd.read_sql(sql, conn)
    
    conn.close()
    
    return df_anuenews

# 讀取【新聞】資料庫
def 連接到xlsx新聞搜索():
    xlsx_path = os.path.join('file', "新聞搜索.xlsx")
    df_news = pd.read_excel(xlsx_path, sheet_name=0)
    
    return df_news

def 篩選鉅亨網公告():
    df_anuenews = 連接到鉅亨網資料庫()
    df_news = 連接到xlsx新聞搜索()
    filtered_df = df_news[~df_news['新聞內容'].isin(df_anuenews['content'])]
    
    filtered_df = filtered_df.reset_index(drop=True)

    return filtered_df

df_news = 篩選鉅亨網公告()
df_news
```

```python
df_news_buy = df_news[df_news['代碼'] == 2330]
df_news_buy = df_news_buy[['代碼', '日期']]

# 2317鴻海所有買進的時間 
df_2317_buy = df_news_buy.sort_values(by = '日期', ascending=True)


# 連接到每日的收盤價
def 連接到收盤價():
    conn = sqlite3.connect(os.path.join('file', 'data.db'))
    sql = "SELECT stock_id, date, 開盤價, 最高價, 最低價, 收盤價 FROM price WHERE stock_id = 2330"
    df_2317 = pd.read_sql(sql, conn)
    
    conn.close()
    
    return df_2317

df_2317 = 連接到收盤價()
df_2317 = df_2317.sort_values(by= 'date', ascending=True)
df_2317
```

# 2317回測

```python
from datetime import datetime

capital = 0
shares = 1
count = 0

for i, row in df_2317_buy.iterrows():
    entry_date = row['日期']
    try:
        entry_price = df_2317[df_2317['date'] == f'{entry_date}']['收盤價'].iloc[0]
        sell_price = 0
        
        # 查找購買日期對應的索引
        entry_index = df_2317[df_2317['date'] == f'{entry_date}'].index.values[0] # 將進場價作為index的第一個
        entry_index = entry_index+1
        
        # 開始遍歷購買日期之後的數據
        for j in range(entry_index, len(df_2317)):
            current_price = df_2317.loc[j, '最高價']
            
            
            # 判斷是否滿足賣出條件（漲幅大於等於2.48%)
            if (current_price >= entry_price * 1.0248).any():
                sell_price = current_price
                sell_date = df_2317.loc[j, 'date']
                break
            
        if sell_price > 0:
            # 计算购买金额和持仓数量
            buy_amount = entry_price * shares
            sell_amount = sell_price * shares
            
            # 天數
            day1 = pd.Timestamp(f'{entry_date}')
            day2 = pd.Timestamp(f'{sell_date}')
            
            delta = day2 - day1
            days_diff = delta.days
            
            # 计算卖出金额和剩余资本
            capital = capital + (sell_amount - buy_amount)
            capital = round(capital, 2)
            
            count += 1
            
            print(f"進場日期: {entry_date}")
            print(f"出場日期: {sell_date}")
            print(f"天數: {days_diff}")
            print(f"進場價格: {entry_price}")
            print(f"售出價格: {sell_price}")
            print(f"獲利: {capital}")
            print("-" * 20)
    except:
        pass

print(f'最終成交次數: {count}')
print(f"最終獲利: {capital}")
```
